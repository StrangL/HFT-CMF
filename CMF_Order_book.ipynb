```python
def statisctics(df_column):
  print('quantile for 0.05 for volume of best ask = ', df_column.quantile(0.05))
  print('quantile for 0.95 for volume of best ask = ', df_column.quantile(0.95))
  print('Mean volume of best ask = ', df_column.mean())
  print('Median volume of best ask = ', df_column.median())
  print('Variance of volume of best ask = ', df_column.var())
```

```python
statisctics(bts_best_ask_volume['btcusdt:Binance:LinearPerpetual_ask_vol_0'])
```

```python
statisctics(btc_lobs['btcusdt:Binance:LinearPerpetual_bid_vol_0'])
```

```python
fig = go.Figure()
for i in range(10):
  fig.add_trace(go.Histogram(x=btc_bid_volumes.iloc[:100000, i].values, name=btc_bid_volumes.columns[i]))

fig.update_yaxes(type="log")

fig.update_layout(title_text='bid volume',
                  barmode='stack',
                  xaxis_title_text='Value',
                  yaxis_title_text='Count')
fig.show()
```

```python
btc_breaking = btc_lobs.loc[:, ['exchange_ts', 'btcusdt:Binance:LinearPerpetual_ask_price_0', 'btcusdt:Binance:LinearPerpetual_ask_vol_0',
                         'btcusdt:Binance:LinearPerpetual_bid_price_0', 'btcusdt:Binance:LinearPerpetual_bid_vol_0']]
btc_breaking = btc_breaking.merge(btc_trades, how="inner", on='exchange_ts')
btc_breaking.sort_values(by='exchange_ts', inplace=True)
btc_breaking.head()
```

```python
btc_ask = btc_breaking[btc_breaking['aggro_side'] == 'ASK']
btc_bid = btc_breaking[btc_breaking['aggro_side'] == 'BID']

btc_group_ask = btc_ask.groupby(by='exchange_ts').agg({'btcusdt:Binance:LinearPerpetual_ask_price_0': 'max', 'price': 'min', 'size' : 'sum'})
btc_group_bid = btc_bid.groupby(by='exchange_ts').agg({'btcusdt:Binance:LinearPerpetual_bid_price_0': 'min', 'price': 'max', 'size' : 'sum'})
```

```python
btc_group_ask['differ'] =  btc_group_ask['price'] - btc_group_ask['btcusdt:Binance:LinearPerpetual_ask_price_0']
btc_group_bid['differ'] = btc_group_bid['price'] - btc_group_bid['btcusdt:Binance:LinearPerpetual_bid_price_0']
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=btc_group_ask['size'].values, y=btc_group_ask['differ'].values, mode='markers', name='ask'))
fig.add_trace(go.Scatter(x=btc_group_bid['size'].values, y=btc_group_bid['differ'].values, mode='markers', name='bid'))

fig.update_layout(title_text='breakout depends on trade size',
                  barmode='stack',
                  xaxis_title_text='size',
                  yaxis_title_text='breakout')
fig.show()
```
